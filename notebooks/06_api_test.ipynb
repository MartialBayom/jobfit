````python
import requests
import os

API_URL = 'http://127.0.0.1:8001'
base    = os.path.expanduser('~/WORKSPACE /PROJETS FIN FORMATION JEDHA/NOUVEAU TAFF/jobfit_phase2')
CV_PATH = os.path.join(base, 'data/cv/sample_cv.pdf')

print('Test de connexion...')
````

## 2. Health check

````python
r = requests.get(f'{API_URL}/health')
print(r.status_code, r.json())
````

## 3. Stats globales

````python
r = requests.get(f'{API_URL}/stats')
import json
print(json.dumps(r.json(), indent=2, ensure_ascii=False))
````

## 4. Scorer le CV contre toutes les offres

````python
with open(CV_PATH, 'rb') as f:
    r = requests.post(
        f'{API_URL}/score/cv-vs-offres',
        files={'cv_file': ('cv.pdf', f, 'application/pdf')},
        params={'top_n': 10}
    )

result = r.json()
print(f'CV : {result["cv_titre"]}')
print(f'Compétences : {result["nb_competences"]}')
print('\nTop 10 offres :')
import pandas as pd
pd.DataFrame(result['top_offres'])[['intitule','entreprise','lieu','type_contrat','ml_score_pct']]
````

## 5. Mode copier-coller — scorer une offre externe

````python
offre_externe = {
    'intitule':     'Data Scientist NLP',
    'description':  'Développement modèles NLP, fine-tuning BERT, déploiement FastAPI Docker AWS MLflow.',
    'entreprise':   'Société Générale',
    'lieu':         'Paris (75)',
    'type_contrat': 'CDI',
    'competences':  'Python, NLP, BERT, PyTorch, Docker, AWS, MLflow',
}

import json

with open(CV_PATH, 'rb') as f:
    r = requests.post(
        f'{API_URL}/score/offre-texte',
        files={'cv_file': ('cv.pdf', f, 'application/pdf')},
        data={'offre_json': json.dumps(offre_externe)}
    )

print(r.status_code)
print(r.text[:500])
````
